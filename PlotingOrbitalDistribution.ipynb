```
import matplotlib.pyplot as plt
import numpy as np
```

```
def plotPoint(r, theta, color = None):
  #print(r*np.cos(theta), r*np.sin(theta))
  if(color==None):
    plt.plot(r*np.cos(theta), r*np.sin(theta), "ro",markersize=0.8)
  else:
    plt.plot(r*np.cos(theta), r*np.sin(theta), color)
```

```
import pymc3 as pm
import random

pdf = pm.Gamma.dist(3, 2/.55)

np.e**(pdf.logp(.55).eval())
```

```
plt.figure(figsize=(10,10))
plt.xlim(-2,2)
plt.ylim(-2,2)
thetas=[]
for i in range(10000):
  thetas.append(random.random()*2*np.pi)
pontos = pdf.random(size=10000)
plt.plot(0,0,"ko")
for i in range(10000):
  plotPoint(pontos[i],thetas[i])
fig
```

```
linha = np.linspace(0,10,100000)
distribution=[]
for x in linha:
  distribution.append(np.e**(-3.63636*x)*(-1 - 3.63636*x - 6.61157*x*x)+1)
print(len(distribution))
```

```
i=1
raios=[]
for x in range(len(linha)):
  if distribution[x]>0.02*i:
    raios.append(linha[x])
    i+=1
```

```
print(raios)
```

```
fig, ax = plt.subplots(figsize=(10,10))

plt.xlim(-3,3)
plt.ylim(-3,3)

for raio in raios:
  ax.add_artist(plt.Circle((0,0),raio,fill=False))
ax.add_artist(plt.Circle((0,0),.55,fill=False, color='r'))
fig.savefig("distribution.png")
```

```
fig, ax = plt.subplots(figsize=(10,10))
plt.xlim(-3,3)
plt.ylim(-3,3)

for i in range(10000):
  ax.plot(pontos[i]*np.cos(thetas[i]), pontos[i]*np.sin(thetas[i]), "ro",markersize=0.8)
for raio in raios:
  ax.add_artist(plt.Circle((0,0),raio,fill=False))
ax.add_artist(plt.Circle((0,0),.55,fill=False, color='orange'))
ax.plot(0,0,"ko")
fig.savefig("orange.png")
```
